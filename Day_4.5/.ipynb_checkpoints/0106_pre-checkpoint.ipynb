```python
'''
q8 写一个程序，逼近圆周率
hint: 圆心到圆上任一点的距离为半径，利用模拟的方式，通过单位圆和长度为1的正方形，可逼近圆周率
'''
```

```python
def pai():
    n = 1
    while True:
        n += 1
        jiaodu = 90 / (2 ** (n+1))
        hudu = jiaodu * 3.14 / 180
        print(hudu)
        bianchang = math.sin(hudu) * 2
        print(bianchang)
        zongchang = bianchang * (2 ** n)
        yield zongchang * 2
```

```python
a = pai()
print(next(a))
```

```python
def ziranduishu():
    n = 0
    while True:
        n += 1
        yield (1 + 1/n) ** n
    
```

```python
def youmeidu(shu):
    fenshu = str.split(str(shu), sep = '0')
    he = 0
    n = 0
    for i in range(len(fenshu)):
        for j in range(len(fenshu[i])):
            he = he + int(fenshu[i][j])
            n += 1        
        if n > 3:
            print('this is not a youmeishu',shu)
            break
    print (he) 
```
