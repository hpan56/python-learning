```python
import pandas as pd
import numpy as np
from functools import reduce
from itertools import combinations
```

```python
ab = {}
ab["金"] = pd.Series({"a":5,"d":3,"r":3,"s":4,"m":4,"h":3})
ab["木"] = pd.Series({"a":3,"d":3,"r":5,"s":3,"m":4,"h":4})
ab["水"] = pd.Series({"a":2,"d":5,"r":4,"s":5,"m":3,"h":3})
ab["火"] = pd.Series({"a":4,"d":3,"r":3,"s":4,"m":5,"h":3})
ab["土"] = pd.Series({"a":2,"d":4,"r":4,"s":5,"m":2,"h":5})
```

```python
df_ab = pd.DataFrame(ab).T
```

```python
ke_dic = {}
ke = ["金", "木", "土", "水", "火", "金"]
for i in range(5):
    ke_dic[ke[i]+ke[i+1]] = "k"
    ke_dic[ke[i+1]+ke[i]] = "-k"
    
sheng_dic = {}
sheng = ["金", "水", "木", "火", "土", "金"]
for i in range(5):
    sheng_dic[sheng[i]+sheng[i+1]] = "s"
    sheng_dic[sheng[i+1]+sheng[i]] = "-s"
```

```python
def rel(a,b):
    try:
        rel_ = ke_dic[a+b]
    except:
        rel_ = sheng_dic[a+b]
    if len(rel_) == 2:
        return b,a,rel_[-1]
    elif len(rel_) == 1:
        return a,b,rel_
    else:
        print("Wrong!")
```

```python
def comp(*arg):
    m,v,r = arg[0][0],arg[0][1],arg[0][2]
    if r == "k":
        if m > v:
            return m+1
        elif m < v:
            return m-1
        else:
            return m
    elif r == "s":
        if m > v:
            return m+2
        elif m < v:
            return m
        else:
            return m+1
    else:
        print("Wrong Relationship given!")
```

```python
def agg(a,b):
    da,db = df_ab.loc[a,],df_ab.loc[b,]
    master,vice, rel_ = rel(a,b)
    df = pd.DataFrame({master:df_ab.loc[master,], vice:df_ab.loc[vice,], "status":rel_})
    df[master+vice] = df.apply(comp, axis=1)
    return df.T.loc[master+vice,]
```

```python
def agg_more(*arg):
    if len(arg) == 2:
        return agg(arg[0],arg[1])
    else:
        temp = pd.DataFrame({})
        for i in list(combinations(arg, len(arg)-1)):
            temp_ = agg_more(*i).astype("int64")
            temp = temp.append(temp_)
        temp = temp[list("adrsmh")]
        temp.drop_duplicates(inplace=True)
        temp.loc["".join(list(arg)),:] = temp.apply(max,axis=0)
        return temp.astype(int)
```

```python
def bonus(*arg):
    temp = arg[0]
    bonus_ls = list(temp.name)
    level = (2 ** len(bonus_ls) - 2) ** 4
    temp += level / 2
    for i in bonus_ls:
        temp[bonus_dic[i]] += round(level,0)
        if i in ["木", "水"]:
            temp["s"] += round(level/2,0) 
        elif i in ["金", "火"]:
            temp["s"] += round(level/4,0) 
    return temp
```

```python
def agg_more_t(*arg):
    if len(arg) == 2:
        return agg(arg[0],arg[1])
    else:
        temp = pd.DataFrame({})
        for i in list(combinations(arg, len(arg)-1)):
            temp_ = agg_more(*i).astype("int64")
            temp = temp.append(temp_)
        temp = temp[list("adrsmh")]
        temp.drop_duplicates(inplace=True)
        temp.loc["".join(list(arg)),:] = temp.apply(sum,axis=0)
        return temp.iloc[-1:,:] * len(arg)
```

```python
final = df_ab.append(agg_more("金", "木", "土", "水", "火"))
bonus_dic = {"金":'a', "木":'r', "水":'d', "火":'m', "土":"h"}
final = final.apply(bonus,axis=1)
final["total"] = final.apply(sum,axis=1)
final = final.astype(int)
final = final.loc[sorted(final.index, key=lambda x: len(x)),:]
final
```
